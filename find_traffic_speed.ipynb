```python
def distance(point1, point2):

    #Radius of earth at latitude 42.89 + elevation of buffalo, ny
    radius = 6368.276 + 0.183

    lat1, lng1 = point1
    lat2, lng2 = point2

    theta1 = math.radians(lat1)
    theta2 = math.radians(lat2)
    phi1 = math.radians(lng1)
    phi2 = math.radians(lng2)

    distance = 2 * radius * math.asin(math.sqrt(math.sin((theta2-theta1)/2)**2 + math.cos(theta1) * math.cos(theta2) * math.sin((phi2-phi1)/2)**2))

    return distance #I'm using kilometers right now
```

```python
def find_nearest_way(datapoint):

    proximal_nodes = {}
    #Determined a +- value of .006 just by looking at the largest space between nodes
    margin = .006

    for key,value in final_node_table.items():
        if value[0] < datapoint[0] + margin and value[0] > datapoint[0] - margin:
            if value[1] < datapoint[1] + margin and value[1] > datapoint[1] - margin:
                proximal_nodes.update({key:value})
    
    proximal_ways = {}
    for key,value in final_way_table.items():
        for node in value:
            if node in proximal_nodes:
                proximal_ways.update({key:value})
    
    
    
    point1 = datapoint
    min_dist = 9999
    min_way = -1
    min_projection = []
    for key,value in proximal_ways.items():
        for i in range(len(value)-1):
            a = final_node_table[value[i]]
            b = final_node_table[value[i+1]]
            c = point1

            u = [b[0]-a[0], b[1]-a[1]]
            v = [c[0]-a[0], c[1]-a[1]]

            projection = [0,0]
            projection[0] = (u[0]*v[0] + u[1]*v[1]) / (u[0]**2 + u[1]**2) * u[0] + a[0]
            projection[1] = (u[0]*v[0] + u[1]*v[1]) / (u[0]**2 + u[1]**2) * u[1] + a[1]

            if distance(a,c) < min_dist:
                min_dist = distance(a,c)
                min_way = key
                min_projection = a

            if distance(b,c) < min_dist:
                min_dist = distance(b,c)
                min_way = key
                min_projection = b

            if min(a[0],b[0]) <= projection[0] <= max(a[0],b[0]) and min(a[1],b[1]) <= projection[1] <= max(a[1],b[1]):

                if distance(projection, c) < min_dist:
                    min_dist = distance(projection,c)
                    min_way = key
                    min_projection = projection
    
    
    return min_projection, min_way
```

```python
with open('data/20200730.csv','r') as f:
    f.readline()
    
    raw = f.readline()
    line = raw.split(',')
    lat = float(line[7])
    lng = float(line[8])
    hours = int(line[10][11:13])
    minutes = int(line[10][14:16])
    seconds = int(line[10][17:19])
    total_seconds = int(line[10][11:13]) * 3600 + int(line[10][14:16]) * 60 + int(line[10][17:19])
    interval = math.floor(total_seconds / 300)
    projection, way = find_nearest_way([lat,lng])
    
    raw = f.readline()
    while raw != '':
        
        previous_lat = lat
        previous_lng = lng
        previous_hours = hours
        previous_minutes = minutes
        previous_seconds = seconds
        previous_total_seconds = total_seconds
        previous_interval = interval
        previous_projection = projection
        previous_way = way
        
        line = raw.split(',')
        lat = float(line[7])
        lng = float(line[8])
        
        if lat == 99.0 and lng == 999.0:
            #I believe this is missing data
            lat = previous_lat
            lng = previous_lng
            #reset these values and skip line
            continue
        
        #print(line)
        hours = int(line[10][11:13])
        minutes = int(line[10][14:16])
        seconds = int(line[10][17:19])
        #print('{}:{}:{}'.format(hours,minutes,seconds))
        total_seconds = int(line[10][11:13]) * 3600 + int(line[10][14:16]) * 60 + int(line[10][17:19])
        interval = math.floor(total_seconds / 300)
        #print(interval)
        #print('{},{}'.format(lat,lng))
        
        projection, way = find_nearest_way([lat,lng])
        
        
        
        
        #km/s -> mph
        #print('{} - {}: {}'.format(previous_total_seconds,total_seconds,raw[:-1]))
        if previous_total_seconds == total_seconds:
            raw = f.readline()
            continue
        if len(previous_projection) == 0:
            print('previous: {}'.format([previous_lat,previous_lng]))
        if len(projection) == 0:
            print('current: {}'.format([lat,lng]))
        speed = distance(previous_projection,projection) / (previous_total_seconds - total_seconds)
        speed = abs(speed)
        #speed = distance([previous_lat,previous_lng],[lat,lng]) / (previous_total_seconds - total_seconds)
        #speed = speed / 1.60934 * 3600
        
        meta_speeds[previous_way][previous_interval].append(speed)
        meta_speeds[previous_way][interval].append(speed)
        meta_speeds[way][previous_interval].append(speed)
        meta_speeds[way][interval].append(speed)
        
        
        #meta_speeds[previous_interval].append(speed)
        #meta_speeds[interval].append(speed)
        
        raw = f.readline()
        #print(raw[:-1])
```
