```python
proximal_nodes = {}
#Determined a +- value of .006 just by looking at the largest space between nodes
margin = .006

for key,value in final_node_table.items():
    if value[0] < datapoint[0] + margin and value[0] > datapoint[0] - margin:
        if value[1] < datapoint[1] + margin and value[1] > datapoint[1] - margin:
            proximal_nodes.update({key:value})
    #elif value[1] < datapoint[1] + margin and value[1] > datapoint[1] - margin:
        #proximal_nodes.update({key:value})
```

```python
proximal_ways = {}
for key,value in final_way_table.items():
    for node in value:
        if node in proximal_nodes:
            proximal_ways.update({key:value})
```

```python
from math import radians, cos, sin, asin, sqrt

def distance(point1, point2):

    #Radius of earth at latitude 42.89 + elevation of buffalo, ny
    radius = 6368.276 + 0.183

    lat1, lng1 = point1
    lat2, lng2 = point2

    theta1 = radians(lat1)
    theta2 = radians(lat2)
    phi1 = radians(lng1)
    phi2 = radians(lng2)

    distance = 2 * radius * asin(sqrt(sin((theta2-theta1)/2)**2 + cos(theta1) * cos(theta2) * sin((phi2-phi1)/2)**2))

    return distance #I'm using kilometers right now
```

```python
point1 = datapoint

min_dist = 9999
min_way = -1
min_projection = []

for key,value in proximal_ways.items():
    for i in range(len(value)-1):
        a = final_node_table[value[i]]
        b = final_node_table[value[i+1]]
        c = point1
        
        u = [b[0]-a[0], b[1]-a[1]]
        v = [c[0]-a[0], c[1]-a[1]]
        
        projection = [0,0]
        projection[0] = (u[0]*v[0] + u[1]*v[1]) / (u[0]**2 + u[1]**2) * u[0] + a[0]
        projection[1] = (u[0]*v[0] + u[1]*v[1]) / (u[0]**2 + u[1]**2) * u[1] + a[1]
        
        if distance(a,c) < min_dist:
            min_dist = distance(a,c)
            min_way = key
            min_projection = a
        
        if distance(b,c) < min_dist:
            min_dist = distance(b,c)
            min_way = key
            min_projection = b
        
        if min(a[0],b[0]) <= projection[0] <= max(a[0],b[0]) and min(a[1],b[1]) <= projection[1] <= max(a[1],b[1]):
            
            if distance(projection, c) < min_dist:
                print(distance(projection,c))
                min_dist = distance(projection,c)
                min_way = key
                min_projection = projection

m = folium.Map(location=point1, tiles="OpenStreetMap", zoom_start=13)
folium.Marker(point1).add_to(m)
folium.Marker(min_projection, popup='projection', icon=folium.Icon(color='red',icon_color='#FFFF00')).add_to(m)

way = []
for node in proximal_ways[min_way]:
    #print(distance(final_node_table[node],point1))
    folium.Marker(location=final_node_table[node]).add_to(m)
    way.append(final_node_table[node])
folium.PolyLine(way).add_to(m)
print('min: {}'.format(min_dist))
m
```
